## Modules

```python
#Generate Random Column Size Changes

def gen_update_file(columns_df, n_ids, new_size, export_csv=False, seed=None):
    if seed is not None:
        np.random.seed(seed)

    n_ids = min(int(n_ids), len(columns_df))
    selected_ids = np.random.choice(columns_df["column_id"].values, size=n_ids, replace=False)

    updates_df = pd.DataFrame({"column_id": selected_ids, "size": new_size})

    if export_csv:
        out_path = os.path.join(os.getcwd(), "change_columns.csv")
        updates_df.to_csv(out_path, index=False)
        print(f"wrote: {out_path}")  # so you can see exactly where it went
        return None  # prevents Jupyter from showing a df as output

    return updates_df
```
